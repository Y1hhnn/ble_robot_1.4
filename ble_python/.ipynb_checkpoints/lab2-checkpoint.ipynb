## ICM Sensor

```python
times = []
accX = []
accY = []
accZ = []
pitch = []
roll = []

def accl_notification_handler(sender, data: bytearray):
    s = ble.bytearray_to_string(data).split("|")
    if len(s) == 6:
        t = float(s[0][2:])
        x = float(s[1][2:])
        y = float(s[2][2:])
        z = float(s[3][2:])
        p = float(s[4][2:])
        r = float(s[5][2:])
        times.append(t)
        accX.append(x)
        accY.append(y)
        accZ.append(z)
        pitch.append(p)
        roll.append(r)
        print (f"x={x:7.2f}   y={y:7.2f}   z={z:7.2f}   pitch={p:7.2f}°   roll={r:7.2f}°")
```

```python
ble.start_notify(ble.uuid["RX_STRING"], accl_notification_handler)
```

```python
ble.stop_notify(ble.uuid["RX_STRING"])
```
